# 실습

```python
import numpy as np

a1 = np.array([1, 5, 2, 10, 8, 9, 300])
print(type(a1))
print(a1)
```

```python
print("배열의 shape: ", a1.shape)
print("배열의 (원소들) 타입: ", a1.dtype)
print("원소의 총 개수: ", a1.size)
print(len(a1))
print("차원 수(rank): ", a1.ndim)
```

```python
# dtype 변경
a2 = a1.astype('int8')

print(a1.dtype, a2.dtype)
print(a1)
print(a2)
```

```python
l = [
    [1, 2, 3],
    [4, 5, 6]
]
a3 = np.array(l, dtype='float32')
print(a3.shape)
print(a3.dtype)
print(a3)
```

```python
l2 = [
    [1, 2, 3],
    [10.0, 20.0, 5.7]
]
a4 = np.array(l2)
print(a4.dtype)
print(len(a4), a4.size)
print(a4)
```

## 동일한 값들로 구성된 배열 생성

### zeros()

```python
a1 = np.zeros(shape = 5)
print(a1)
print(a1.dtype)
```

```python
a2 = np.zeros(shape=(5, 4, 3), dtype = 'float32')
print(a2.shape, a2.dtype)
print(a2)
```

### ones()

```python
a3 = np.ones(shape=(4, 2))
print(a3.shape, a3.dtype)
print(a3)
```

### full()

```python
a4 = np.full(shape=(3, 4), fill_value = 'a')
print(a4.shape, a4.dtype)
print(a4)
```

```python
a5 = np.zeros_like(a4)
a6 = np.ones_like(a4)
a7 = np.full_like(a4, fill_value = 100) # 기존 배열이 문자열일때 type과 문자열개수에 맞춰진다.
print(a5.shape, a6.shape, a7.shape)
print(a5)
print(a6)
print(a7)
```

## 특정 범위내에서 동일한 간격의 값들로 구성된 배열 생성

### arange(start, stop, step, dtype)

```python
a1 = np.arange(1, 10, 2)
a2 = np.arange(-10, 10, 2)
print(a1)
print(a2)
```

```python
a3 = np.arange(0, 1.1, 0.1)
print(a3)
```

```python
a4 = np.arange(10, 0, -2)
a5 = np.arange(10, 20)
a6 = np.arange(20)

print(a4)
print(a5)
print(a6)
```

### linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None)

```python
a7 = np.linspace(1, 100)
a8 = np.linspace(1, 100, 1000)
print(a7.shape, a8.shape)
print(a7)
print(a8)
```

```python
a9 = np.linspace(1, 100, 5, retstep = True, endpoint = False)
print(a9)
```

```python
def f(x):
    return x**2 + 2*x + 3
```

```python
x = np.linspace(-5, 5, 1000)
y = f(x)
print(y)
```

```python
import matplotlib.pyplot as plt
plt.plot(x, y)
plt.show()
```

## 난수(Random value)를 원소로 하는 ndarray 생성

### np.random.seed(시드값)

```python
np.random.seed(212) 
# 난수 발생 함수 알고리즘에 처음 들어가는 수를 0으로 지정
# -> 난수함수를 발생하면 일정한 난수 값이 나온다. 
```

```python
np.random.randint(1, 100, size = (4, 5))
```

### np.random.rand([axis0, axis1, axis2, ...])

```python
print(np.random.rand())
print("=" * 30)
print(np.random.rand(5))
print("=" * 30)
print(np.random.rand(3, 2))
print("=" * 30)
print(np.random.rand(3, 4, 2))
```

### np.random.normal(loc=0.0, scale=1.0, size=None)

```python
a = np.random.normal()
# 평균: 0, 표준편차 : 1 => 표준 정규 분포
# 2표준편차 범위(평균 - 2*표준편차 ~ 평균 + 2*표준편차) => 95%
a
```

```python
a2 = np.random.normal(100, 20)
a2
```

```python
a3 = np.random.normal(10, 2, size = (5, 3))
print(a3.shape)
a3
```

```python
plt.hist(a3.flatten())
plt.show()
```

```python
plt.hist(np.random.normal(10, 2, size = 10000), bins = 30)
plt.show()
```

### np.random.randint(low, high=None, size=None, dtype='int32')

```python
print(np.random.randint(100))
print("=" * 30)
print(np.random.randint(-10, 10))
print("=" * 30)
print(np.random.randint(-10, 10, size = 5))
print("=" * 30)
print(np.random.randint(-10, 10, size = (2, 3, 4)))
print("=" * 30)
print(np.random.randint(10, size = (3, 2)))
```

### np.random.choice(a, size=None, replace=True, p=None)

```python
print(np.random.choice([1, 2, 3], size = 10))
print(np.random.choice([1, 2, 3], size = (3, 2)))
```

```python
np.random.choice([1, 2, 3], size = 10, replace = False) # 중복추출이 안됨(replace = False) ==> 비복원 추출
                                                        # 모수의 size 초과 -> 추출 안됨
```

```python
a = np.random.choice([True, False], size = (5, 5), p = [0.9, 0.1])
a
```
